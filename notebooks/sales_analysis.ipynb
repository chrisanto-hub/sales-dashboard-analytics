```python
# Changing the data type of discounted price and actual price
# Changing the data type of discounted price and actual price

data['discounted_price'] = data['discounted_price'].str.replace("₹",'')
data['discounted_price'] = data['discounted_price'].str.replace(",",'')
data['discounted_price'] = data['discounted_price'].astype('float64')

data['actual_price'] = data['actual_price'].str.replace("₹",'')
data['actual_price'] = data['actual_price'].str.replace(",",'')
data['actual_price'] = data['actual_price'].astype('float64')
```

```python
# Changing Datatype and values in Discount Percentage
data['discount_percentage'] = data['discount_percentage'].str.replace("%",'')
data['discount_percentage'] = data['discount_percentage'].astype('float64')
data['discount_percentage'] = data['discount_percentage'] / 100.0
```

```python
# Find missing values percentage in the data
missing_values = data.isnull().sum()
percentage_missing = (missing_values / len(data)) * 100
percentage_missing
```

**plot the missing values by percentage**

```python
# plot the missing values by percentage
plt.figure(figsize=(10,6))
sns.barplot(x=percentage_missing.index, y=percentage_missing.values, palette='viridis')
plt.xticks(rotation=45)
plt.ylabel('Percentage of Missing Values')
plt.title('Missing Values Percentage by Column')
plt.show()
```

```python
#view rows with null values in rating_count column
data[data['rating_count'].isnull()]
data[data['rating_count'].isnull()]['rating'].mean()
# Impute missing values in rating_count with mean of rating where rating_count is null
mean_rating_count = data[data['rating_count'].isnull()]['rating'].mean()
data['rating_count'] = data['rating_count'].fillna(mean_rating_count)

```

**Grouping and Aggregation**

```python
#Convert product_name → CATEGORY
data['category'] = data['product_name'].apply(lambda x: x.split()[0] if pd.notnull(x) else 'Unknown')
data['category'].value_counts()

# Calculate mean rating by category
mean_rating_by_category = data.groupby('category')['rating'].mean()
mean_rating_by_category
```

**summary statistics for groups**

```python
#summary statistics for groups
# Mean rating by category
mean_rating_by_category = data.groupby('category')['rating'].mean()
mean_rating_by_category
```

**Create pivot tables**

```python
#Create pivot tables
pivot_table = data.pivot_table(values='discounted_price', index='category', columns='rating', aggfunc='mean')
pivot_table
```

**Statistical Tests:**

```python
# Conduct t-test to compare rating between two categories
from scipy.stats import ttest_ind

# Select ratings for two categories
category1_ratings = data[data['category'] == 'Electronics']['rating']
category2_ratings = data[data['category'] == 'Clothing']['rating']

# Perform t-test
t_stat, p_value = ttest_ind(category1_ratings, category2_ratings)
t_stat, p_value
```

**Chi-square test**

```python
# Chi-square test
from scipy.stats import chi2_contingency
# Create a contigency table
contingency_table = pd.crosstab(data['category'], data['rating'])
chi2, p, dof, expected = chi2_contingency(contingency_table)
chi2, p
```
